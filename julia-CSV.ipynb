```julia
# 
x = 0:0.01*pi:2*pi
y = sin.(x)
d = Dict("x"=>x,"y"=>y)
df = DataFrame(d)

# CSV 
# CSV.write("sin.csv",df)
# df = CSV.read("sin.csv")

# 
fig, ax = subplots(figsize=(9,6))
f1 = plot(df.x,df.y)
plot(df.x,df.y .^2)
plot(df.x,df.y .^3)

# 
legend(["sin(x)","sin(x)","sin(x)"],
    loc="best",
    bbox_to_anchor=(0.5,0.5,0.5,0.5),
    fontsize=17)
# 
xlim([0,2*pi])
# 
xtick = 0:1/2:2
xticks(xtick .* ,string.(180 .* xtick).*" ")
# 
# xlabel("x")
# ylabel("y")
# 
title(L"y = sin(x)-sin(x)")
sinx = L"sin(x)"
# 
# annotate("max",xy=(,0),xytext=(,0.1))
# annotate(f1,[(x[10],y[10],text(sinx,16,:center))])
# 
grid(true,which="major",axis="both",color="c")
# 
box(false)
```
