```julia
using CSV
using DataFrames
using PyPlot
using LaTeXStrings
```

```julia
# 数据定义
x = 0:0.01*pi:2*pi
y = sin.(x)
d = Dict("x"=>x,"y"=>y)
df = DataFrame(d)

# CSV 文件读写
# CSV.write("sin.csv",df)
# df = CSV.read("sin.csv")

# 初等绘图
fig, ax = subplots(figsize=(9,6))
f1 = plot(df.x,df.y)
plot(df.x,df.y .^2)
plot(df.x,df.y .^3)

# 图例
legend(["sin(x)","sin²(x)","sin³(x)"],
    loc="best",
    bbox_to_anchor=(0.5,0.5,0.5,0.5),
    fontsize=17)
# 坐标轴范围
xlim([0,2*pi])
# 坐标轴刻度
xtick = 0:1/2:2
xticks(xtick .* π,string.(180 .* xtick).*" ᴼ")
# 坐标轴标题
# xlabel("x")
# ylabel("y")
# 图标题
title(L"y = sin(x)-sin²(x)")
sinx = L"sin(x)"
# 注释
# annotate("max",xy=(π,0),xytext=(π,0.1))
# annotate(f1,[(x[10],y[10],text(sinx,16,:center))])
# 网格线
grid(true,which="major",axis="both",color="c")
# 外边框
box(false)
```
