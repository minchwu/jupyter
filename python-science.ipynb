```python
import numpy as np
import matplotlib as mpt
from matplotlib import ticker
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,np.pi,46)
y1 = np.sin(x)*np.exp(x)
y2 = np.cos(x)*np.power(x,2)
with plt.style.context(['science','no-latex']):
#     fig, ax = plt.subplots()
    fig = plt.figure(figsize=[9,6],dpi=100)
    ax = fig.add_axes([0.1,0.1,0.5,0.5])
#     ax.plot(x,y1,x,y2)
#     ax.legend(title='Func',labels=('sinx','cosx'),loc=0)
    ax.plot(x,y1,linestyle='-.',linewidth=2,color='#FFAA00',label='$e^xsinx$')
    ax.plot(x,y2,linestyle='-',linewidth=2,color='#DD0000',marker='.',label='$x^2cosx$')
    ax.legend(title='Func',loc=0,frameon=True,shadow=True)
    ax.grid(b=True,color='#ABCDEF')
    ax.autoscale(tight=True)
#     plt.xticks(ticks=np.linspace(0,np.pi,8),labels=['π/8','π/4','3π/8','π/2','5π/8','3π/4','7pi/8','π'])
    plt.xticks(ticks=np.linspace(0,np.pi,4),labels=['π/4','π/2','3π/4','π'])

#     fig.savefig('./sin.png')
```

```python
x = np.linspace(0, 2 * np.pi, 361)
y = np.exp(np.sin(x))
z = np.sin(np.exp(x))
with plt.style.context(['science', 'ieee', 'no-latex']):
    fig = plt.figure(num=1, figsize=(8, 6), dpi=100)
    fig.suptitle("$x-y-z$")
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.5])
    ax.plot(x, y, linestyle='-.', linewidth=2, color='r', label='$exp(sinx)$')
    ax.legend(frameon=True, shadow=True, loc=3)
    ax.set_xlabel('x')
    ax.grid(b=True, color='c')
    ax.autoscale(tight=True)
    bx = fig.add_axes([0.1, 0.7, 0.4, 0.2])
    bx.plot(x, z, linewidth=1, color='b', label='$sin(exp(x))$')
    bx.legend(frameon=True, shadow=True, fontsize=7, loc=3)
    cx = fig.add_axes([0.6, 0.7, 0.3, 0.2])
    cx.plot(y, z, linewidth=1, color='k', label='$y-z$')
    cx.legend(frameon=True, shadow=True, fontsize=7, loc=3)
#     plt.show(fig)
```

```python
x = [0, 0.2]
y = [0, 395702.389]
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,2))
with plt.style.context(['science','ieee','no-latex']):
    fig, ax = plt.subplots()
    fig.suptitle("Amplitude-Time(N)")
    ax.plot(x,y,color='b',marker='o',markersize=3,markerfacecolor='k',label="Amp(N)")
    ax.legend()
    ax.set_xlabel('time/s')
    ax.set_ylabel('amplitude/N')
#     ax.set_yscale('log')
    ax.yaxis.set_major_formatter(formatter)
    plt.xticks([0,0.2])
#     plt.yticks([0,395702])
#     plt.autoscale(tight=True)
    fig.savefig('bclutch.svg')
```
