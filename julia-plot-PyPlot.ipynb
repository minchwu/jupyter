## 正弦函数测试

```julia
# 数据定义
x = 0:0.01*pi:2*pi
y1 = sin.(x)
y2 = 2*sin.(2 .*x)
y3 = 3*sin.(3 .*x)

# 图框设置
figure(1,figsize=(9,6))
box(false)
grid(true)

# 坐标轴设置
title("y = Asin(Bx)")
# xlabel("x")
# ylabel("y")
xlim([0,2*pi])
ylim([-3.1,3.1])
xticks(0:1/2*pi:2*pi,["0","1/2π","π","3/2π","2π"])
label = ("sinx","2sin2x","3sin3x")

plot(x,y1,"r-",x,y2,"g-",x,y3,"b-")
legend(label,loc="best",fontsize=17)

# savefig("AsinBx.png")
```
